```python
# this tool takes in a FASTA file of dna and translates it directly into codons and their respective amino acids.
# this block contains the main() function used as an entry point to the whole code and its multiple functions.
def main():
    filename = input("enter sequence filename: ")
    sequence = read_file(filename)
    codons = get_codons(sequence)
    aa_list = trans_codons(codons)
    
    print("stripped sequence:",sequence)
    print("the codons are: ",codons)
    print("the amino acids are: ",aa_list)

if __name__ == "__main__":
    main()
```

```python
# this function reads the file line by line, strips any space,comma or unwanted noise in the sequence and puts it into a string.
def read_file(filename):
    with open(filename,'r') as file:
        lines = file.readlines()
        
        
        sequence = " "
        
        for line in lines:
            if not line.startswith(">"):
                sequence += line.strip()
        return sequence
```

```python
# this function then converts the sequence into triplet codons.
def get_codons(sequence):
    sequence = sequence.strip().upper()
    
    codons = []

    remainder = len(sequence) % 3
    if remainder != 0:
        sequence = sequence[:-remainder]

    for i in range(0,len(sequence),3):
        codon = sequence[i:i+3]
        if len(codon) == 3:
            codons.append(codon)
    return codons    
```

```python
# here a dictionary is made with codons and their respectice amino acids so that the codons are translated to provide protein skeleton by the last function.
genetic_code = {
    "TTT": "F", "TTC": "F",
    "TTA": "L", "TTG": "L",
    "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",

    "ATT": "I", "ATC": "I", "ATA": "I",
    "ATG": "M",

    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",

    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S",
    "AGT": "S", "AGC": "S",

    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",

    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",

    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",

    "TAT": "Y", "TAC": "Y",

    "AAA": "K", "AAG": "K",
    
    "CAT": "H", "CAC": "H",

    "CAA": "Q", "CAG": "Q",

    "AAT": "N", "AAC": "N",


    "GAT": "D", "GAC": "D",

    "GAA": "E", "GAG": "E",

    "TGT": "C", "TGC": "C",

    "TGG": "W",

    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGA": "R", "AGG": "R",

    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",

    "TAA": "*", "TAG": "*", "TGA": "*"
}

def trans_codons(codons):
    
    aa_list = []

    for codon in codons:
        amino_acids = genetic_code[codon]
        aa_list.append(amino_acids)
    return aa_list    
    
```
